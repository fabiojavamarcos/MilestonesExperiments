# Likert Plot Analysis

```python

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = './data/final3.csv'
data = pd.read_csv(file_path)
    
```

```python

# Select and clean the Likert scale questions
likert_questions = data.dropna()

# Count responses for each Likert scale option
likert_counts = likert_questions.apply(pd.Series.value_counts).fillna(0).astype(int)

# Define Likert scale order and reorder data
likert_order = ["Strongly disagree", "Somewhat disagree", "Neither agree nor disagree", "Somewhat agree", "Strongly agree"]
likert_counts = likert_counts.reindex(likert_order)
    
```

```python

# Plot the Likert data
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(8, 10), sharex=True)

for idx, question in enumerate(likert_counts.columns):
    axes[idx].barh(likert_order, likert_counts[question])
    axes[idx].set_title(question)

plt.suptitle("Likert Responses for Each Question")
plt.xlabel("Number of Responses")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('eval-likert-task.png', format='png')
plt.savefig('eval-likert-task.pdf', format='pdf')

plt.show()
    
```
