# Likert Plot Analysis - Separate Plots

```python

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = './data/final3.csv'
data = pd.read_csv(file_path)
    
```

```python

# Select and clean the Likert scale questions
likert_questions = data.dropna()

# Count responses for each Likert scale option
likert_counts = likert_questions.apply(pd.Series.value_counts).fillna(0).astype(int)

# Define Likert scale order and reorder data
likert_order = ["Strongly disagree", "Somewhat disagree", "Neither agree nor disagree", "Somewhat agree", "Strongly agree"]
likert_counts = likert_counts.reindex(likert_order)
    
```

```python

# Plot for question: Do the labels make you feel more confident to contribute to this project?
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(likert_order, likert_counts['Do the labels make you feel more confident to contribute to this project?'])
ax.set_title("Do the labels make you feel more confident to contribute to this project?")
ax.set_xlabel("Number of Responses")
plt.tight_layout()
plt.savefig('eval-likert-project.png', format='png')
plt.savefig('eval-likert-project.pdf', format='pdf')
plt.show()
    
```

```python

# Plot for question: Do the labels make you feel more confident to contribute to a task?
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(likert_order, likert_counts['Do the labels make you feel more confident to contribute to a task?'])
ax.set_title("Do the labels make you feel more confident to contribute to a task?")
ax.set_xlabel("Number of Responses")
plt.tight_layout()



# Save the plot in PNG and PDF formats
plt.savefig('eval-likert-tasks.png', format='png')
plt.savefig('eval-likert-tasks.pdf', format='pdf')
plt.show()    
```

```python

# Plot for question: Do the labels clearly explain the skills you need to contribute to a task?
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(likert_order, likert_counts['Do the labels clearly explain the skills you need to contribute to a task?'])
ax.set_title("Do the labels clearly explain the skills you need to contribute to a task?")
ax.set_xlabel("Number of Responses")
plt.tight_layout()
plt.savefig('eval-likert-explaination.png', format='png')
plt.savefig('eval-likert-explaination.pdf', format='pdf')
plt.show()
    
```

```python

# Plot for question: Are the labels relevant as a source of information about a task you want to contribute to
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(likert_order, likert_counts['Are the labels relevant as a source of information about a task you want to contribute to'])
ax.set_title("Are the labels relevant as a source of information about a task you want to contribute to")
ax.set_xlabel("Number of Responses")
plt.tight_layout()
plt.savefig('eval-likert-source.png', format='png')
plt.savefig('eval-likert-source.pdf', format='pdf')
plt.show()
    
```
